```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("ratings_small.csv")
```

```python
df1.head(4)
```

```python
df1['id']=df1['movieId']
```

```python
df1 = df1[["userId","id","rating"]]
```

```python
df1.head(4)
```

```python
df2 = pd.read_csv("movies_metadata.csv", low_memory=False)
```

```python
df2.head(4)
```

```python
df1["id"] = pd.to_numeric(df1["id"], errors="coerce").astype("Int32")
df2["id"] = pd.to_numeric(df2["id"], errors="coerce").astype("Int32")
```

```python
df1 = df1.dropna(subset=["id"])
df2 = df2.dropna(subset=["id"])
```

```python
new_df = pd.merge(df1, df2, on="id", how="inner")
```

```python
new_df
```

```python
new_df.columns
```

```python
new_df = new_df[["userId","id","rating","imdb_id","title"]]
```

```python
new_df.head(4)
```

```python
new_df.isnull().sum()
```

```python
new_df = new_df.drop(["id"], axis = 1)
```

```python
new_df.head(6)
```

```python
user_rating = new_df.pivot_table(index = ['userId'], columns = ['title'], values='rating')
```

```python
user_rating
```

```python
user_rating = user_rating.fillna(0)
```

```python
user_rating.head(5)
```

```python
def standardize(row):
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row
    
```

```python
user_rating = user_rating.apply(standardize)
```

```python
user_rating.head(6)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_similarity = cosine_similarity(user_rating.T)
```

```python
movie_similarity
```

```python
movie_similarity_df = pd.DataFrame(movie_similarity, index = user_rating.columns, columns = user_rating.columns)
```

```python
movie_similarity_df.head(6)
```

```python
movie_similarity_df.shape
```

```python
def recommend(movie_name,rating):
    similar_score=movie_similarity_df[movie_name]*(rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    similar_score = list(similar_score.index)
    return similar_score[0:10]
```

```python
recommend("Zombie Holocaust",3)
```
